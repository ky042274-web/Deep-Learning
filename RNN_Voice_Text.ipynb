```python
import speech_recognition as sr
```

```python
# Initialize recognizer
recognizer = sr.Recognizer() #IT IS A CONTROLLER WHICH LISTEN TO AUDIO AND CONVERT TO TEXT
```

```python
with sr.Microphone() as source: #IT open the microphone as the audio source
    recognizer.adjust_for_ambient_noise(source, duration=1)
    print("Speak now:")
    audio = recognizer.listen(source) #records voice untill we stop speaking

try:
    text = recognizer.recognize_google(audio, language="en-IN,hi-IN")  # It sends the recorded audio to Google's Speech Recognition
    print("Text:", text)

    text = recognizer.recognize_google(audio, language="hi-IN")  # It sends the recorded audio to Google's Speech Recognition
    print("Hindi Text:", text)

except sr.UnknownValueError:
    print("Could not understand")
```
