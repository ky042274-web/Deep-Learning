```python
from transformers import MarianTokenizer, MarianMTModel

model_name = "Helsinki-NLP/opus-mt-en-hi"
```

```python
# Load tokenizer and model
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
# Text to translate
text = ["The Ganga (Ganges) is the sole officially declared National River of India, designated as such on November 4, 2008, to represent the country's national symbols. It is a 2,525 km long river originating from the Gangotri Glacier, revered as a sacred, life-nurturing mother river spanning across Northern and Eastern India"]
```

```python

# Tokenize text
tokens = tokenizer(text, return_tensors="pt", padding=True)

# Generate translation
translated = model.generate(**tokens)

# Decode and print
output = tokenizer.decode(translated[0], skip_special_tokens=True)
print("Translated:", output)
```
