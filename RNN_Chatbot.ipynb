```python
def chatbot():
    print("Komal: Hi there! Welcome to NSTI Allahabad CITS Portal")

    while True:
        user = input("You: 🫵 ").lower()

        if "hi" in user or "hello" in user:
            print("NSTI: Hey there! How may I help you?")

        elif "where" in user and "located" in user:
            print("NSTI: It is located at 6, New Katra, Prayagraj.")

        elif "courses" in user:
            print("NSTI: We offer COPA, CSA, AIAPA, EM, OM, COSMO, DM, FDT, C&H.")

        elif "copa" in user and "admission" in user:
            print("NSTI: For COPA admission, you must have completed ITI in COPA trade.")

        elif "eligibility" in user or "qualification" in user:
            print("NSTI: Eligibility: ITI in COPA or equivalent qualification.")

        elif "duration" in user:
            print("NSTI: COPA course duration is 1 year.")

        elif "fees" in user:
            print("NSTI: Fees are as per government norms and are minimal.")

        elif "bye" in user:
            print("NSTI: Bye! Take care 👋")
            break

        else:
            print("NSTI: Sorry, I didn't understand that. Please ask about COPA admission, eligibility, duration or fees.")

# Run chatbot
chatbot()
```
