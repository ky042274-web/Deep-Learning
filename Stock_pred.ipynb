```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("tata_motors_ohlc.CSV")
```

```python
data
```

```python
prices=data['Close'].values.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()  #it scales values into range (0 to 1)
price_scaled=scaler.fit_transform(prices) #it learn the min /max from our data and transform accordingly
```

```python
#creating sequenser last 5 Day > Next Day

X=[]
y=[]

for i in range(len(price_scaled)-5):
    X.append(price_scaled[i:i+5])  #5 Days previous prices features
    y.append(price_scaled[i+5]) #the next value (Target) 

X=np.array(X)
y=np.array(y)
```

```python
#Build RNN Model
from tensorflow .keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense
from tensorflow.keras.layers import Input, LSTM,Dense
```

```python
#model build
model = Sequential()

model.add(LSTM(50, input_shape=(5, 1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
# Train the model
model.fit(X, y, epochs=15, batch_size=16, validation_split=0.1)
```

```python
#predictionNext Day price

# Take last 5 days from scaled data
five_day = price_scaled[-5:]          # shape (5,1)

# Reshape to (1, timesteps, features) for model input
five_day = five_day.reshape(1, 5, 1)   # 1 sample, 5 timesteps, 1 feature

prediction = model.predict(five_day)
predicted_price=scaler.inverse_transform(prediction)

print ("Prediction of the Next Day close price is:",predicted_price)
```
